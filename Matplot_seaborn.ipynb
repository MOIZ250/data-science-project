```python
import pandas as pd
file_name="Iris.csv"
df=pd.read_csv(file_name)

df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python

x = df["SepalLengthCm"]
y = df["SepalWidthCm"]
```

```python

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Fig1.png')
```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'y*',linestyle='dashed',linewidth=2, markersize=9)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'g--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*--')
plt.subplot(2,2,3)
plt.plot(x,y,'ro')
plt.subplot(2,2,4)
plt.plot(x,y,'y--')
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi)
y = np.sin(x)
plt.title("sine wave form")

# Plot the points using matplotlib
plt.plot(x, y)
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin,'r--')
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos,'g--')
plt.title('Cosine')

# Show the figure.
plt.show()
```

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y)
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
